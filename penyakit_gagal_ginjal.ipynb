# Metode Crips DM

# 1. Business Understanding

Business Understanding adalah bertujuan untuk memahami masalah bisnis :

Masalah : Masalah nya adalah Penderita penyakit ginjal terus meningkat, Gagal Ginjal terjadi karena kerusakan ginjal sudah cukup berat atau berlangsung lama sehingga penyakit ini umumnya merupakan tahapan akhir dari penyakit ginjal. penyebab terjadi nya gagal Ginjal adalah Volume aliran darah ke ginjal yang rendah, Saluran kemih yang terganggu, Memiliki diabetes, Tekanan darah tinggi, Mengonsumsi makanan dan minuman yang tidak terjaga, Terlalu sering menahan buang air kecil, dan Pola hidup tidak sehat.

Tujuan : Tujuan nya adalah untuk mengidentifikasi apakah seseorang itu mengidap menyakit gagal Ginjal atau tidak. yang nanti nya digunakan untuk membantu bidang kesehatan dan mengurangi beban dokter, dan juga untuk mengembangkan program pencegahan terjadi nya penyakit gagal Ginjal agar menjadi lebih efektif.


# 2. Data Understanding

Data Understanding ini berisikan kondisi kesehatan Penyakit Gagal Ginjal apakah status seseorang itu mengidap menyakit gagal Ginjal atau tidak.

Mendeskripsikan 400 Orang dan 14 Atribut

Isi:

1.Bp : Bp atau Blood Pressure adalah tekanan darah

2.Sg : Sg atau Specific Gravity adalah berat badan

3.Al : Al atau Albumin adalah protein pada darah yang membentuk sebagian besar plasma darah.

4.Su : Su atau Sugar adalah gula pada tubuh, karbohidrat yang merupakan substrat energi utama tubuh

5.Rbc : Rbc atau Red Blood Cell adalah Sel darah merah pada tubuh, bagian sel darah yang mengandung hemoglobin

6.Bu : Bu atau Blood Urea adalah urea darah pada tubuh, zat sisa dari pemecahan protein dan asam amino di dalam hati

7.Sc : Sc atau Serum Creatinine adalah Kreatinin serum, sampah hasil metabolisme otot yang mengalir pada sirkulasi darah

8.Sod : Sod atau Sodium adalah mineral penting, Sodium membantu tubuh kita agar bisa berfungsi dengan baik, misalnya memungkinkan otot dan saraf berfungsi dengan benar

9.Pot : Pot atau Pottasium adalah Kalium, mineral dalam tubuh yang mengendalikan fungsi sel saraf dan otot, terutama otot jantung

10.Hemo : Hemo atau Hemoglobin adalah protein yang terdapat pada sel darah merah

11.Wbcc : Wbcc atau White Blood Cell Count adalah Sel Darah Putih, Leukosit atau sel darah putih merupakan bagian dari sistem kekebalan tubuh yang berfungsi untuk melindungi diri dari infeksi atau penyakit

12.Rbcc : Rbcc atau Red Blood Cell Count adalah Jumlah Sel Darah Merah

13.Htn : Htn atau Hypertension adalah Hipertensi

14.Class : Class atau Predicted Class adalah status orang tersebut apakah mengidap penyakit gagal Ginjal atau tidak, Jika mengidap berstatus 1 dan jika tidak mengidap bersatus 0

# 4. Modeling

# Seleksi Fitur

x : blood_pressure, sg, albumin, sugar, red_blood_cell, blood_urea, serum_creantinine, sodium, pottasium, hemoglobin,
    white_blood_cell,red_blood_cell_count, hypertension
    
y : class

```python
#Memisahkan Fitur dan Label
features = ['blood_pressure','sg','albumin','sugar','red_blood_cell','blood_urea','serum_creatinine',
            'sodium','pottasium','hemoglobin','white_blood_cell','red_blood_cell_count','hypertension']
x = df[features]
y = df['class']
x.shape, y.shape
```

```python
#Menampilkan x dan y
#x adalah Atribut (blood_pressure, sg, albumin, sugar, red_blood_cell, blood_urea, serum_creantinine, sodium, 
#pottasium, hemoglobin, white_blood_cell,red_blood_cell_count, hypertension)
#y adalah Label (class)
print(x)
print(y)
```

# Memisahkan Data Training dan Data Testing

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Logistic Regression Model

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

# 5. Evaluation

# Model Evaluasi Logistic Regression Model

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Akurasi data training adalah: ',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Akurasi data testing adalah: ',test_data_accuracy)
```
